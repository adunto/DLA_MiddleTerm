```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler
import pandas as pd
```

```python
default_cycler = (cycler(color=['r', 'g', 'b', 'y']) +
                  cycler(linestyle=['-', '--', ':', '-.']))
plt.rc('lines', linewidth=2)
plt.rc('axes', prop_cycle=default_cycler)
```

```python
file_paths = [".result/DNN_result.csv", ".result/CNN_result.csv",
               ".result/ResNet_result.csv", ".result/ResNet_Custom_result.csv"]
labels = ["DNN", "CNN", "ResNet", "ResNet(Customized)"]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle("Training Metrics by Model", fontsize=16)

for file_path, label in zip(file_paths, labels):
    data = pd.read_csv(file_path)
    
    ax1.plot(data['epoch'], data['accuracy'], label=label)
    ax2.plot(data['epoch'], data['loss'], label=label)

ax1.set_title("Epoch : Accuracy")
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Accuracy")
ax1.legend()
ax1.grid(True)

ax2.set_title("Epoch : Loss")
ax2.set_xlabel("Epoch")
ax2.set_ylabel("Loss")
ax2.legend()
ax2.grid(True)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
file_path = [".result/ResNet_result.csv", ".result/ResNet_Custom_result.csv"]
labels = ["ResNet", "ResNet(Customized)"]

fig, ax = plt.subplot(figsize=(16, 8))
fig.suptitle("Original vs Customized", fontsize=16)

for file_path, label in zip(file_paths, labels):
    data = pd.read_csv(file_path)
    
    ax.plot(data['epoch'], data['accuracy'], label=label)
    ax.plot(data['epoch'], data['loss'], label=label)
```
